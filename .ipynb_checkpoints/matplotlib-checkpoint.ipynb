```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,50)
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x),x,np.sin(2*x))#画两个图
```

```python
plt.plot(x,np.sin(x),'r-o')
plt.plot(x,np.cos(x),'g--')#定制化线条
```

# Colours:
- Blue – ‘b’
- Green – ‘g’
- Red – ‘r’
- Cyan – ‘c’
- Magenta – ‘m’
- Yellow – ‘y’
- Black – ‘k’ (‘b’ is taken by blue so the last letter is used)
- White  – ‘w’
# Lines:
- Solid Line – ‘-‘
- Dashed – ‘–‘
- Dotted – ‘.’
- Dash-dotted – ‘-:’
# Often Used Markers:
- Point – ‘.’
- Pixel – ‘,’
- Circle – ‘o’
- Square – ‘s’
- Triangle – ‘^’

```python
#subplot 允许在一个窗口放置多张图
x=np.linspace(0,2*np.pi,50)
plt.subplot(2,1,1)#(row,column,active area)
plt.plot(x,np.sin(x),'r.')
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'bo')
```

```python
y=np.sin(x)
plt.scatter(x,y)#散点图
```

```python
x=np.random.rand(1000)
y=np.random.rand(1000)
size=np.random.rand(1000)*50
colour=np.random.rand(1000)
```

```python
plt.scatter(x,y,size,colour)#color map scatter，By doing this, the points plotted on the graph will vary in size and colour depending on the data we pass.
plt.colorbar()#colorbar
```

```python
x=np.random.randn(1000)
```

```python
plt.hist(x,50)#直方图，第二个参数指定bin的个数
```

```python
x=np.linspace(0,2*np.pi,50)
plt.plot(x,np.sin(x),'r-x',label='Sin(x)')#label
plt.plot(x,np.cos(x),'g-^',label='Cos(x)')
plt.legend()#lengend
plt.xlabel('Rads')#xlabel
plt.ylabel('Amplitude')#ylabel
plt.title('Sin and Cos Waves')#title
```
